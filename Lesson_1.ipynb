### __1. Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.__

```python
import math

def comb(n, k):
    return int(math.factorial(n) / (math.factorial(k) * math.factorial(n - k)))

def accom(n, k):
    return int(math.factorial(n) / math.factorial(n - k))
```

$A=a_1+a_2+a_3+a_4$ - число исходов, соответствующих тому, что был вытянут хотя бы один туз из 4 вытянутых карт из колоды 52 карт, где

$a_1$ - число исходов вытянуть ровно $1$ туз и $3$ другие карты,

$a_2$ - число исходов вытянуть ровно $2$ туза и $2$ другие карты,

$a_3$ - число исходов вытянуть ровно $3$ туза и $1$ другие карты,

$a_4$ - число исходов вытянуть ровно $4$ туза.

$a_i$ - это произведение сочетаний вытянуть $n$ тузов из $4$ и сочетания вытянуть оставшиеся карты (до $4$) не тузы.

```python
A = comb(4, 1) * comb(48, 3) + comb(4, 2) * comb(48, 2) + comb(4, 3) * comb(48, 1) + comb(4, 4) * comb(48, 0)
print(f'Число искомых исходов: {A}')
```

### __2. Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что рядом будут сидеть: а) два определённых человека? б) три определённых человека?__

$P(A)=\frac{m}{n}$, где n - количество благоприятных (искомых) исходов, m - общее число исходов

$n$ - произведение исходов, что $2 (3)$ определенных человека сядут рядом и перестановок $5 (4)$ человек по оставшимся $5 (4)$ местам

Количество вариантов двух определенных людей рядом = ($7-1$ (крайнее место)) $*2!$ (количество перестановок двух людей)
Количество вариантов трех определенных людей рядом = ($7-2$ (крайнее место для двух человек)) $*3!$ (количество перестановок трех людей)

$m$ - общее число исходов = общему кол-ву перестановок $7$ человек = $7!$

```python
P_a = (6*math.factorial(2)*math.factorial(5))/math.factorial(7)
P_b = (5*math.factorial(3)*math.factorial(4))/math.factorial(7)
print(f'Вероятности:\nа) {P_a}\nb) {P_b}')
```

### __5. Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.__
### __Вопрос: какова вероятность, что вы действительно больны ей?__
### __Подсказка: вновь используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.__

Пусть $A$ - болен, а $B$ - тест положительный

Тогда по формуле Байеса:
$P(A|B)=\frac{P(B|A)*P(A)}{P(B)}$

А из формулы полной вероятности^ $P(B)=P(B|A)*P(A)+P(B|\bar A)*P(\bar A)$

Тогда: $P(A|B)=\frac{P(B|A)*P(A)}{(P(B|A)*P(A)+P(B|\bar A)*P(\bar A))}$

$P(A)=0,001$

$P(B|A)=0,99$

```python
p = (0.99*0.001)/(0.99*0.001+0.01*(1-0.001))
print(f'Вероятность болезни: {p} = {p:%}')
```
