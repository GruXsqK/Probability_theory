# Непрерывные случайные величины. Функция распределения и плотность распределения вероятностей. Равномерное и нормальное распределение. Центральная предельная теорема

__1. О непрерывной равномерно распределённой случайной величине B известно, что её дисперсия равна 0.2. Можно ли найти правую границу величины B и её среднее значение зная, что левая граница равна 0.5? Если да, найдите их.__

Математическое ожидание данной величины вычисляется по формуле:$M(X) = \dfrac{a + b}{2}$, а дисперсия: $D(X) = \dfrac{(b - a)^2}{12}$ и равна она $0.2$, левая граница, т.е. $a=0.5$.

Подставим известные в формулу дисперсии и найдем неизвестные:

$\dfrac{(b - 0.5)^2}{12} = 0.5\Rightarrow (b - 0.5)^2 = 6$

Далее можем раскрыть скобки и решить квадратное уравнение. Но т.к. правая граница больше левой, то $b>0.5\Rightarrow b-0.5>0$, то:

$b-0.5 = \sqrt{6}\Rightarrow b = \sqrt{6} + 0.5$

Зная обе границы можем найти среднее значение, т.е. мат.ожидание

```python
import numpy as np

a = 0.5
b = np.sqrt(6)+0.5
M = (a + b)/2
print(f'Правая граница = {b}\nМатематическое ожидание = {M}')
```

__2. На сколько сигм (средних квадратических отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 и D(X) = 25?__

Для нормального распределения непрерывной случайной величины $M(X) = a, \:\: D(X) = \sigma^2$

Следовательно $\sigma = \sqrt{D(X)} = \sqrt{25}=5$, а отклонение роста: $\Delta = 190 - 178 = 12$

Рост человека, равный $190$ см отклоняется на  $\dfrac{\Delta}{\sigma} = 2.4 $ сигм

__3. Рост взрослого населения города X имеет нормальное распределение, причём, средний рост равен 174 см, а среднее квадратическое отклонение равно 8 см. С помощью правил стандартного отклонения, двух и трёх сигм посчитайте, какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:__

__1) больше 182 см?__

__2) больше 190 см?__

__3) от 166 см до 190 см?__

__4) от 166 см до 182 см?__

__5) от 158 см до 190 см?__

__6) не выше 150 см или не ниже 190 см?__

__7) не выше 150 см или не ниже 198 см?__

__8) ниже 166 см?__

Все расчеты будем производить исходя из следующего:

Средний рост взрослого населения: $a = 174$

Среднее квадратичное отклонение: $\sigma = 8$

Вероятность попасть в интервал от  $166$ до $182$ (стандартное отклонение): $P(1\sigma) = 0.6826894921370859$

Вероятность попасть в интервал от  $158$ до $190$: $P(2\sigma) = 0.9544997361036416$

Вероятность попасть в интервал от  $150$ до $198$: $P(3\sigma) = 0.9973002039367398$

1) $P(x>182) = \dfrac{1 - P(1\sigma))}{2}\quad$- т.к. мы ищем вероятность роста вне диапазона $1\sigma$ и только с одной стороны этого диапазона

2) $P(x>190) = \dfrac{1 - P(2\sigma)}{2}$

3) $P(166<=x<=190) = \dfrac{P(1\sigma)}{2}+\dfrac{P(2\sigma)}{2}$

4) $P(166<=x<=182) = P(1\sigma)$

5) $P(158<=x<=190) = P(2\sigma)$

6) $P(150=>x \& x>=190) = \dfrac{1 - P(3\sigma)}{2} + \dfrac{1 - P(2\sigma)}{2}$

7) $P(150=>x \& x>=198) = 1 - P(3\sigma)$

8) $P(x<166) = P(x>182) = \dfrac{1 - P(1\sigma)}{2}$

```python
from scipy import stats

a = 174
sig = 8
p_sig_1 = stats.norm.cdf(1) - stats.norm.cdf(-1)
p_sig_2 = stats.norm.cdf(2) - stats.norm.cdf(-2)
p_sig_3 = stats.norm.cdf(3) - stats.norm.cdf(-3)
print(f"""Вероятности определенного роста случайно выбранного человека
1) {(1-p_sig_1)/2}
2) {(1-p_sig_2)/2}
3) {p_sig_1/2+p_sig_2/2}
4) {p_sig_1}
5) {p_sig_2}
6) {(1-p_sig_3)/2 + (1-p_sig_2)/2}
7) {1-p_sig_3}
8) {(1-p_sig_1)/2}""")
```
